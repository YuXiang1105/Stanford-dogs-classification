```python
from datasets import load_dataset
from PIL import Image
from io import BytesIO
```

```python
#the source of this dataset is hugging face, by amaye15/stanford-dogs,
#link: https://huggingface.co/datasets/amaye15/stanford-dogs
df = load_dataset("amaye15/stanford-dogs")
train_df = df['train'].to_pandas()
test_df = df['test'].to_pandas()
```
